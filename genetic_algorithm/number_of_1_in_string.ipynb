```python
import pygad
import numpy as np

initial_population = np.random.randint(2, size=(100, 20))


def fitness_func(ga_instance, solution, solution_idx):
    chromosome = solution

    # converting binary array to int value
    x = np.dot(chromosome, 2 ** np.arange(len(chromosome))[::-1])

    return x


fitness_function = fitness_func

ga_instance = pygad.GA(
    gene_type=int,
    initial_population=initial_population,
    num_generations=200,
    num_parents_mating=50,
    fitness_func=fitness_function,
    sol_per_pop=100,
    num_genes=len(initial_population),
    mutation_type="random",
    mutation_probability=0.001,
    crossover_type="single_point",
    crossover_probability=1
)

ga_instance.run()

print("Best solution ", ga_instance.best_solution())

ga_instance.plot_fitness()
```
